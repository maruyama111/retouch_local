# 詳細設計書

## 1. システムの概要
本システムは、ローカル環境で高速かつ安定的な写真レタッチを可能にするアプリケーションです。
以下の主要な機能を通じて、初心者でも直感的に利用できる操作性を実現します。

### 目的
- ローカル環境での高速写真処理（1枚あたり1秒以下）。
- ネットワーク依存の解消により、安定した処理環境を提供。
- 大量の写真（100枚以上）を効率的に一括処理。
- 初心者ユーザー向けのシンプルかつ直感的な操作性。

### 対象機能
1. モデルの追加学習機能  
   - レタッチ前後の写真を使用して既存モデルを拡張。  
2. バッチ処理機能  
   - 選択したモデルを使い、複数の写真に一括でレタッチ処理を適用。  
3. バッチ処理の枚数制限設定  
   - ユーザーが処理可能な写真の最大枚数を設定し、性能と操作性を両立。

## 3. シーケンス

### 3.1 モデルの追加学習
- ユーザーがローカル環境からモデルを選択し、レタッチ前後の写真セットを登録。
- 学習が完了すると、更新されたモデルを保存。

### 3.2 バッチ処理
- ユーザーがモデルと写真を選択し、バッチ処理を実行。
- 処理が完了した写真は一括保存。

### 3.3 バッチ処理の枚数制限設定
- 現在の制限値を確認し、新しい制限値を入力。
- 設定内容を保存。
