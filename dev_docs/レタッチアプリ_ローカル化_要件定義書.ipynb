# 要件定義

## 1. 機能の概要
### 1.1 機能の目的

本プロジェクトの目的は、写真レタッチアプリをローカル環境化することで、以下を実現することです。
- 処理速度の向上（1セットの写真処理時間を1秒以下に短縮）。
- 大量写真処理の効率化（100枚以上のバッチ処理対応）。
- ネットワーク依存の解消による安定性の向上。
- 初心者でも直感的に操作可能なUI/UXの提供。

### 1.2 機能の対象範囲
- 対象範囲は既存Webアプリの機能を拡張・改善し、ローカル環境での使用を想定した高速処理機能を中心に開発します。

### 1.3 機能の利用者
- 一般ユーザー：簡易的な写真編集を求める初心者ユーザー。

## 2. 機能要件
### 2.1 対象機能
- ローカル保存のモデル選択：ローカル環境内の保存済みモデルを直接選択し、登録する機能。ネットワーク依存を解消し、高速処理を実現。
- バッチ処理機能：100枚以上の写真を一括でレタッチ処理できる機能。大量の写真を効率的に処理するための高速処理対応。
- ローカルフォルダ選択機能：ローカルフォルダから直接写真を読み込んで処理する機能。アップロード待機時間を削減し、作業効率を向上。
- バッチ処理の枚数制限設定：ユーザーが一度に処理する写真の最大枚数を設定可能。パフォーマンスと操作性を両立する。


### 2.2 対象画面
- トップページ
  - モデルの追加学習ページ（追加学習版アプリ）
    - モデルの登録
    - レタッチ前後の写真セットの登録
    - ローカル保存のモデル選択機能（新規追加）
  - 新しい画像へのモデルの適用ページ（アウトプット版アプリ）
    - モデルの登録
    - レタッチ前の写真の登録
    - バッチ処理機能（新規追加）
    - ローカルフォルダ選択機能（新規追加）
  - 設定画面（新規追加）
    - バッチ処理の枚数制限の設定

## 3. 非機能要件
### 3.1 性能
- 処理速度：1セットの写真処理時間を1秒以下に短縮。
- バッチ処理対応：100枚以上の写真を一括で処理可能。

### 3.2 操作性
- UI/UXの直感性：初心者でも簡単に操作可能なインターフェースを設計。
- エラーメッセージの明確化：処理失敗時に具体的で分かりやすいエラーメッセージを表示。
